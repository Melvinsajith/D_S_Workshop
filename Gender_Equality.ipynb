```python
ymp_pred = model.predict(xmp_test)
ymp_pred
```

```python
#FEMALE POPULATION
columns_to_extract_F= [	'Population_Female',	'Population_Female2']

extracted_data_F = extracted_data.loc[::, columns_to_extract_F]

xfp = extracted_data_F.iloc[:,:-1]
yfp = extracted_data_F.iloc[:,-1]
```
